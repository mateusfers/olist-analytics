```python
import pandas as pd
import numpy as np

df_orders = pd.read_csv(r'C:\Users\Mateus Fernandes\Documents\olist-analytics\data\raw\olist_orders_dataset.csv')

print("Dados carregados com sucesso!!! Verificação!")
print(f"Linhas: {df_orders.shape[0]}")
print(f"Colunas: {df_orders.shape[1]}")
print(f"Valores tupla: {df_orders.shape}")
```

```python
df_orders['order_purchase_timestamp'] = pd.to_datetime(df_orders['order_purchase_timestamp'])
df_orders['order_delivered_customer_date'] = pd.to_datetime(df_orders['order_delivered_customer_date'])
df_orders['order_estimated_delivery_date'] = pd.to_datetime(df_orders['order_estimated_delivery_date'])

print("Colunas convertidas!!")

# Verificar os tipos
print(df_orders[['order_purchase_timestamp', 'order_delivered_customer_date']].dtypes)

```

```python
# Calcular a diferença em dias
df_orders['delivery_time'] = (df_orders['order_delivered_customer_date'] - df_orders['order_purchase_timestamp']).dt.days

# Ver as primeiras linhas com a nova coluna
df_orders[['order_purchase_timestamp', 'order_delivered_customer_date', 'delivery_time']].head()
```

```python
print("=== ESTATÍSTICAS DO TEMPO DE ENTREGA ===")
print(df_orders['delivery_time'].describe())

print(f"\nValores nulos em delivery_time: {df_orders['delivery_time'].isnull().sum()}")
```
